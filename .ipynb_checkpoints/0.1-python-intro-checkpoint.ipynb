# Fluxo de controle e condicionais

```python
for i in range(5):
    print(i)
```

```python
for i in range(5):
    if i%2 == 0:
        print('  par %d' % i)
    else:
        print('impar %d' % i)
```

```python
for i in range(12):
    if i%2 == 0 and i%4 == 0:
        print('{0:2d} par e multiplo de 4'.format(i))
    elif i%2 == 0:
        print('{:2d} par'.format(i))
    else:
        print('{:2d} impar'.format(i))
```

```python
i = 0
while True:
    i += 1
    if i == 5:
        break
    print(i)
```

# Funções e funções anônimas

```python
def fat(n):
    return 1 if n == 0 else n * fat(n-1)
```

```python
fat(10)
```

```python
f = fat
f(3)
```

```python
def aplique(f, arg):
    return f(arg)
```

```python
aplique(fat, 4)
```

```python
map(f, range(5))
```

```python
map(lambda n: n*n, range(10))
```

```python
filter(lambda n: n%2==0, range(10))
```

```python
reduce(lambda x,y: x+y, range(10))
```

```python
[n*n for n in range(10)]
```

```python
[n for n in range(10) if n%2==0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plotserie(X, title = None, **kws):
    plt.plot(X, **kws)
    if title:
        plt.title(title)
```

```python
plotserie([np.sin(x) for x in np.linspace(-np.pi, np.pi, 30)], title = 'plot', color = 'red')
```

## Classes e objetos

```python
class Pessoa(object):
    "Representa uma pessoa que tem nome e fala"
    n_pessoas = 0
    
    def __init__(self, n):
        self.nome = n
        Pessoa.n_pessoas += 1
        
    def fala(self):
        print('Meu nome eh', self.nome)
        
    def __str__(self):
        return self.nome
                
class Aluno(Pessoa):
    "Representa um aluno que faz um curso"
    def __init__(self, n, c = 'computacao'):
        Pessoa.__init__(self, n)
        self.curso = c
        
    def fala(self):
        Pessoa.fala(self)
        print('Eu curso', self.curso)
```

```python
a1 = Aluno('Joao', 'Matematica')
a2 = Aluno('Ana', 'Filosofia')
```

```python
a1.fala()
```

```python
a2.fala()
```

```python
Pessoa.n_pessoas
```

```python
print(a2)
```
