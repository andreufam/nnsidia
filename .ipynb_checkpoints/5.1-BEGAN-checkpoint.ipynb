```python
from keras import backend as K
from keras.engine import Input, Model
from keras.engine.topology import Container, Layer
from keras.layers import Lambda, Concatenate, Flatten, Dense, Reshape, Convolution2D, UpSampling2D
from keras.optimizers import Adam
```

Como vimos, redes geradoras adversariais (GANs) são modelos baseados na competição entre duas redes neurais, a geradora $G$ e a discriminadora $D$. Para que o treino da GAN seja bem sucedido, portanto, é necessário que haja um certo equilíbrio entre $G$ e $D$, já que uma depende da outra para aprender. Se $D$ é muito fraco, ele não vai induzir $G$ ao aprendizado, com $G$ possivelmente entrando em colapso; se é muito forte, ele pode não dar qualquer chance para $G$, uma vez que qualquer mudança de estratégia de $G$ continua levando a imagens vistas como falsas por $D$. 

Note que há uma certa contradição nesta estratégia já que, ao fim do treino, queremos o desequilíbrio entre $G$ e $D$, com $G$ vencendo a disputa -- ou seja, $G$ gera imagens que $D$ não consegue mais reconhecer como reais. 

Os casos degenerados (colapsos) são particularmente importantes pelo que nos ensinam sobre o problema. Eles acontecem quando $G$ aprende a enganar $D$ sem apresentar diversidade. Por exemplo, $G$ memoriza imagens que sabe que enganam $D$ e gera apenas elas. Isto é importante porque indica que _um gerador bem sucedido não necessariamente precisa aprender a real distribuição dos dados_.

Muitas heurísticas têm sido usadas para lidar com estes problemas como (i) treinar os modelos de forma alternada para tentar induzir algum equilíbro, (ii) usar modelos menores (ou mais simples) para $G$ de forma que ele não seja tão forte tão cedo (note que o problema a ser resolvido por $G$ é mais complexo que o de $D$ o que, de certa forma, é um motivador para esta estratégia), (iii) usar dropout em $D$ para que ele não aprenda coisas muito específicas que podem ser descobertas por $G$, levando-o ao colapso, (iv) usar regularizadores para desacelerar o aprendizado de $D$, (v) melhorar a função de perda (ver https://arxiv.org/abs/1704.00028, https://arxiv.org/abs/1611.04076 e https://arxiv.org/abs/1701.06264, como exemplos). 

Nesta aula, vamos estudar um outro modelo de GAN, a Boundary Equilibrium Generative Adversarial Network -- BEGAN, que aborda vários dos problemas anteriores de forma menos _ad-hoc_. Em suma, os autores da BEGAN propõem:

* um discriminador que não se restringe a classificar imagens como verdadeiras ou falsas, mas verifica quão efetivamente o gerador está aproximando a distribuição das imagens reais;
* um critério objetivo para medir o equilíbrio entre $D$ e $G$, e o uso  deste critério para balancear o treino de maneira a fortalecer um deles sempre que fica mais fraco que o outro; e 
* uma métrica de convergência capaz de avaliar a convergência do aprendizado conjunto.

### A distância de Wasserstein

Antes de começarmos, vamos ver como comparar diferentes distribuições. Para isso, são dadas três distribuições de idades abaixo:

```python
from scipy.stats import norm
```

```python
data = [np.hstack([np.random.randint(0, 17, 200), # most children
                   np.random.randint(18, 30, 50),
                   np.random.randint(31, 60, 50),
                   np.random.randint(61, 90, 50)]),
        np.hstack([np.random.randint(0, 17, 50),
                   np.random.randint(18, 30, 200), # most young
                   np.random.randint(31, 60, 50),
                   np.random.randint(61, 90, 50)]),
        np.hstack([np.random.randint(0, 17, 50),
                   np.random.randint(18, 30, 50),
                   np.random.randint(31, 60, 50),
                   np.random.randint(61, 90, 200)])] # most eldery
```

```python
dists = []
mu = [0, 0, 0]
std = [0, 0, 0]
fig, axs = plt.subplots(1,3, figsize = (24, 8), sharey = True)
for i in range(len(axs)):
    mu[i], std[i] = norm.fit(data[i])
    dist = axs[i].hist(data[i], bins=25, normed=True, 
             alpha=0.6, color='g')
    dists += [dist[0]] # the bin's (states') probabilities
    x = np.linspace(min(data[i]), max(data[i]), 100)
    p = norm.pdf(x, mu[i], std[i])
    axs[i].plot(x, p, 'k', linewidth=2)
    plt.setp(axs[i].get_xticklabels(), fontsize=22)
    plt.setp(axs[i].get_yticklabels(), fontsize=22)
plt.show()
```

Quais são as mais similares entre si? Como vimos, é possível inferir isso com métricas como a $D_{KL}$, que compara distribuições. 

```python
def DKL(p, q):
    return sum([p[i] * np.log(p[i]/q[i]) for i in range(len(p))])
```

```python
print (DKL(dists[0], dists[1]), 
       DKL(dists[0], dists[2]), 
       DKL(dists[1], dists[2]))
```

De acordo com a $D_{KL}$ não há grande diferença entre as distribuições. Contudo, isso não casa com nossa intuição. Parece óbvio que as duas primeiras distribuições (amostras de crianças e jovens) são mais similares entre si que quando comparadas com a terceira (idosos), não?

Uma outra métrica para comparar distribuições é a distância de Wasserstein. Intuitivamente, se cada distribuição for vista como uma pilha de terra, a métrica é o mínimo custo para transformar uma pilha na outra, estimada como _o montante de terra a ser movido vezes a distância movida_ (essa métrica também é conhecida como _distância do movedor de terra_).

Para duas distribuições Gaussianas em $\mathbb{R}^n$, $N_1 = \mathcal{N}(\mu_1, C_1)$ e $N_2 = \mathcal{N}(\mu_2, C_2)$, com matrizes de co-variância simétricas semi-definidas $C_1$ e $C_2 \in \mathbb{R}^{n \times n}$, a distância de Wasserstein entre $N_1$ e $N_2$ (com respeito a um espaço Euclidiano normalizado) é dada por:

$$W_{2}(N_{1}, N_{2})^{2}=\|\mu_{1}-\mu_{2}\|_{2}^{2}+\mathop {\mathrm {trace} {\bigl (}} C_{1}+C_{2}-2{\bigl (}C_{2}^{1/2}C_{1}C_{2}^{1/2}{\bigr )}^{1/2}{\bigr )}$$

Para $\mathbb{R}^1$ esta equação se reduz a:

$$W_{2}(N_{1},N_{2})^{2}=|\mu_{1}-\mu_{2}|^{2} + C_{1} + C_{2} - 2 \sqrt{C_{1} C_{2}}$$

```python
def wasserstein(N1, N2):
    return np.sqrt(np.abs(N1[0] - N2[0]) + 
                   N1[1] + N2[1] - 2 * np.sqrt(N1[1] * N2[1]))
```

```python
print (wasserstein((mu[0], std[0]), (mu[1], std[1])), 
wasserstein((mu[0], std[0]), (mu[2], std[2])), 
wasserstein((mu[1], std[1]), (mu[2], std[2])))
```

Agora, as diferenças entre as distribuições parecem casar com a nossa intuição. A BEGAN usa esta métrica para comparar distribuições.

### O objetivo da BEGAN (sem equilíbrio)

Como vimos, o discriminador $D$ de uma GAN deveria verificar se $G$ está aproximando a distribuição original das imagens. Uma forma de fazer isso é pensar em $D$ como um autocodificador, tal que a distribuição das imagens reconstruídas por $D$ seja próxima a das imagens reais. Esta ideia foi primeiro sugerida nas energy-based GANs, ou EBGANs, cuja arquitetura é ilustrada abaixo:

<img src="images/ebgan.png" alt="exemplos BEGAN" style="width: 400px;"/>

Note que o erro de reconstrução não é o objetivo final da GAN. No desenho, o objetivo final é L, que será derivado de alguma forma a partir do erro de reconstrução de D. A ideia fundamental da BEGAN, inspirada na EBGAN, é como definir L. Em particular, a BEGAN parte da hipótese de que _casar as distribuições das imagens geradas por G com a dos dados reais é equivalente a casar os respectivos erros de reconstrução_. Logo, a função de perda pode ser derivada da distância de Wasserstein entre os erros de reconstrução das imagens reais e geradas. 

Seja $\mathcal{L}(v) = |v − \mathcal{D}(v)|$ o erro de reconstrução de uma imagem $v$, onde $\mathcal{D}(v)$ é a função auto-codificadora discriminativa $\mathcal{D}: \mathbb{R}^{|v|} \rightarrow \mathbb{R}^{|v|}$. Vamos supor que o erro de reconstrução de $\mathcal{D}$ para uma imagens real $x$ e uma falsa $G(z)$ segue uma distribuição normal, ou seja, $\mathcal{L}(x) \approx \mathcal{N}(\mu_1, C_1)$ e $\mathcal{L}(G(z)) \approx \mathcal{N}(\mu_2,C_2)$. Logo, a distância de Wasserstein entre a reconstrução de uma imagem real e a reconstrução de uma imagem falsa é dado por:

$$W_{2}(\mathcal{L}(x), \mathcal{L}(G(z))^{2}=\|\mu_{1}-\mu_{2}\|_{2}^{2}+C_{1}+C_{2}-2 \sqrt{C_{1}C_{2}}$$

Se $\frac{C_{1}+C_{2}-2 \sqrt{C_{1}C_{2}}}{\|\mu_{1}-\mu_{2}\|_{2}^{2}}$ for constante ou monotônicamente crescente, podemos afirmar que:

$$W_{2}(\mathcal{L}(x), \mathcal{L}(G(z)) \propto \|\mu_{1}-\mu_{2}\|_{2}$$

Note que D quer maximizar a distância entre $\mathcal{L}(x)$ e $\mathcal{L}(G(z))$, uma vez que ele  reconstruir bem imagens reais, mas não imagens falsas. Para maximizar $W_2$, então é necessário minimizar o erro de reconstrução $\mu_1$ e maximizar o erro de reconstrução $\mu_2$. Ou seja, minimizar o erro de reconstrução da imagem real (autocodificar a imagem real) e maximizar o erro de reconstrução da imagem gerada (tratá-la como anomalia). Por sua vez, minimizar o erro do _gerador_ é basicamente minimizar o erro da imagem gerada, de forma que o discriminador não consiga percebê-la como anomalia. Logo, o objetivo da BEGAN se traduz em:

$$
\begin{cases}
\mathcal{L}_D = \mathcal{L}(x; \phi) − \mathcal{L}(G(z_D; \theta); \phi) & \text{for } \phi \\
\mathcal{L}_G = \mathcal{L}(G(z_G; \theta)) & \text{for } \theta
\end{cases}
$$

onde $\phi$ corresponde aos pesos e bias de $D$ e $\theta$, aos pesos e bias de $G$. E a BEGAN é formada por (i) uma rede geradora $G$ e (ii) uma rede auto-codificadora discriminativa $\mathcal{D}$ que reconstrói imagens reais e interpreta imagens geradas como anomalias.

### Introduzindo Equilíbrio

Podemos dizer que o gerador e o discriminador estão em equilíbrio se $\mathbb{E}[\mathcal{L}(G(z))] = \mathbb{E}[\mathcal{L}(x)]$. Esta é uma situação indesejável durante o treino porque indica que D não sabe a diferença entre imagens reais e falsas e, por isso, G não será mais capaz de obter gradientes úteis. Podemos representar o equilíbrio, ao longo do treino, através do coeficiente $\gamma$:

$$\gamma = \frac{\mathbb{E}[\mathcal{L}(G(z))]}{\mathbb{E}[\mathcal{L}(x)]}$$

Valores baixos de $\gamma$ (se afastando de 1 em direção ao 0, ou seja, $\mathcal{L}(G(z)) \gg \mathcal{L}(x)$) indicam que $D$ está sendo capaz de reconhecer as imagens falsas (geradas). Logo, $G$ é incentivado a produzir imagens mais realistas (e, portanto, menos diversas/anômalas) para re-equilibrar a situação. 

Por outro lado, valores altos de $\gamma$ (se aproximando de 1) indicam que imagens falsas estão tão convincentes quantos as reais e, portanto, $D$ deve melhorar a sua capacidade de discriminação, ao passo que $G$ está mais livre para produzir imagens mais diversas, ao custo de possível perda em qualidade.

Mas como usar $\gamma$ para afetar o treino? Se considerarmos que nossa taxa de equilíbrio é $\lambda = \mathbb{E}[\mathcal{L}(G(z))] / \mathbb{E}[\mathcal{L}(x)]$, em certo instante temos $\lambda = \mathcal{L}(G(z)) / \mathcal{L}(x)$, que é equivalente a $\lambda \mathcal{L}(x) - \mathcal{L}(G(z))$ = 0. Esta equação é satisfeita com $\lambda = 1$, o que não queremos de verdade ao longo do treino, pois implica em $D$ incapaz de levar $G$ ao aprendizado. Quando $\lambda \neq 0$, o fator $\lambda \mathcal{L}(x) - \mathcal{L}(G(z))$  indica o quanto estamos afastados do equilíbrio. Logo, $D$ deveria prestar mais ou menos atenção na reconstrução de imagens falsas proporcionalmente a este fator.  

Assim, podemos introduzir $\gamma$ no treino, alterando levemente o objetivo da BEGAN:

$$
\begin{cases}
\mathcal{L}_D = \mathcal{L}(x) − k_t \mathcal{L}(G(z_D)) & \text{for } \phi \\
\mathcal{L}_G = \mathcal{L}(G(z_G)) & \text{for } \theta \\
k_{t + 1} = k_t + \lambda_k (\gamma \mathcal{L}(x) − \mathcal{L}(G(z_G))) & \text{for each training step } t
\end{cases} 
$$

$\gamma$ indica o quanto $G$ e $D$ deveriam ser treinados de forma que um não fique muito melhor que o outro. Agora, $k_t$ controla o quanto $D$ tem que focar em imagens falsas. Ele muda em todos os batches. $k_t$ é cortado para se manter no intervalo entre 0 e 1 e é iniciado em 0, uma vez que o foco inicial deve ser reconstruir imagens reais. $\lambda_k$ é uma espécie de taxa de aprendizado de $k_t$, definida como $0.001$ e, como vimos antes, $k_t$ aumenta ou diminui de acordo com o fator $\gamma \mathcal{L}(x) - \mathcal{L}(G(z_G) = 0$.

### Convergência

Ao contrário de muitas GANs que precisam atualizar $G$ e $D$ de forma independente, a propriedade de equilíbrio controlado das BEGANs as permite treinar a rede como um todo, atualizando $G$ e $D$ simultaneamente (embora seja ainda necessário atualizar os parâmetros com respeito às suas funções de perda específicas):

$$
\DeclareMathOperator*{\argmin}{argmin}
\argmin_{\phi}{\mathcal{L}_D} + \argmin_{\theta}{\mathcal{L}_G}
$$

Além disso, note que à medida que o treino avança e a GAN melhora seu desempenho, $\mathcal{L}(x)$ deve se tornar menor, pois o discriminador é cada vez mais capaz de reconstruir imagens, ao mesmo tempo que o fator $| \gamma \mathcal{L}(x) - \mathcal{L}(G(z_G))|$ se aproxima de zero, uma vez que a rede deve terminar balanceada (com $G$ sendo capaz de produzir saídas cada vez mais consistentes). 

logo, é possível inferir quão bem a rede está aprendendo usando uma métrica de convergência, $\mathcal{M}_{global}$, definida como:

$$\mathcal{M}_{global} = \mathcal{L}(x) + |~ \gamma \mathcal{L}(x) - \mathcal{L}(G(z_G)) ~|$$

Claramente $\mathcal{M}_{global}$ cai se a diferença do erro de reconstrução entre a imagem real e a falsa cai, ou seja, se $G$ se torna um gerador melhor de imagens.

### Criando o modelo

O discriminador da versão da BEGAN que vamos implementar é um autocodificador com codificador e decodificador como ilustrado na figura abaixo. O gerador é basicamente o decodificador. A BEGAN original ainda possui a opção de algumas conexões de _skip_ como na ResNet.

<img src="images/began_model.png" alt="BEGAN model" style="width: 800px;"/>

```python
class BEGANConfig:
    def __init__(self):
        self.image_width = 64
        self.image_height = 64
        self.n_filters = 128     # min number of filters
        self.n_layer_in_conv = 2 # #convs in each "same-size layer"
        self.emb_size = 64       # embedding size
        self.k_0 = 0             # 1st focus is real image reconstruction
        self.gamma = 0.5
        self.lambda_k = 0.001
        self.batch_size = 16
        self.lr_0 = 0.0001       # adaptive learning rate params
        self.min_lr = 0.00001
        self.lr_decay_rate = 0.9
```

```python
def build_model(cfg):
    K.set_image_data_format('channels_last')

    autoencoder = build_autoencoder(cfg)
    generator = build_generator(cfg)
    discriminator = build_discriminator(cfg, autoencoder)

    return autoencoder, generator, discriminator

def build_autoencoder(cfg, name="autoencoder"):
    encoder = build_encoder(cfg, name="%s/encoder" % name)
    decoder = build_decoder(cfg, name="%s/decoder" % name)
    autoencoder = Container(encoder.inputs, 
                            decoder(encoder.outputs), name=name)
    return autoencoder

def build_generator(cfg):
    decoder = build_decoder(cfg, name="generator_decoder")
    generator = Model(decoder.inputs, decoder.outputs, name="generator")
    return generator

def build_encoder(cfg, name="encoder"):
    n_filters = cfg.n_filters
    emb_size = cfg.emb_size
    n_layer = cfg.n_layer_in_conv

    dx = image_input = Input((cfg.image_height, cfg.image_width, 3))

    # output: (N, 32, 32, n_filters)
    dx = conv_enc(dx, n_filters, strides=(2, 2), name="%s/L1" % name, n_layer=n_layer)

    # output: (N, 16, 16, n_filters*2)
    dx = conv_enc(dx, n_filters * 2, strides=(2, 2), name="%s/L2" % name, n_layer=n_layer)

    # output: (N, 8, 8, n_filters*3)
    dx = conv_enc(dx, n_filters * 3, strides=(2, 2), name="%s/L3" % name, n_layer=n_layer)

    # output: (N, 8, 8, n_filters*4)
    dx = conv_enc(dx, n_filters * 4, strides=(1, 1), name="%s/L4" % name, n_layer=n_layer)

    dx = Flatten()(dx)
    hidden = Dense(emb_size, activation='linear', name="%s/Dense" % name)(dx)

    encoder = Container(image_input, hidden, name=name)
    return encoder

def build_decoder(cfg, name):
    """
    generator and decoder( of discriminator) have same network structure, but don't share weights.
    This function takes different input layer, flow another network, and return different output layer.
    """
    n_filters = cfg.n_filters
    n_layer = cfg.n_layer_in_conv

    dx = input_z = Input((64, ))
    dx = Dense((8*8*n_filters), activation='linear', name="%s/Dense" % name)(dx)
    dx = Reshape((8, 8, n_filters))(dx)

    # output: (N, 16, 16, n_filters)
    dx = conv_dec(dx, n_filters, upsample=True, name="%s/L1" % name, n_layer=n_layer)

    # output: (N, 32, 32, n_filters)
    dx = conv_dec(dx, n_filters, upsample=True, name="%s/L2" % name, n_layer=n_layer)

    # output: (N, 64, 64, n_filters)
    dx = conv_dec(dx, n_filters, upsample=True, name="%s/L3" % name, n_layer=n_layer)

    # output: (N, 64, 64, n_filters)
    dx = conv_dec(dx, n_filters, upsample=False, name="%s/L4" % name, n_layer=n_layer)

    # output: (N, 64, 64, 3)
    image_output = Convolution2D(3, (3, 3), padding="same", activation="linear", name="%s/FinalConv" % name)(dx)
    decoder = Container(input_z, image_output, name=name)
    return decoder

def build_discriminator(cfg, autoencoder):
    """
    Keras Model class is able to have several inputs/outputs. 
    But, loss functions should be defined each other, and the loss function cannot reference other inputs/outputs.
    For computing loss, two inputs/outputs are concatenated.
    """
    # IN Shape: [ImageHeight, ImageWidth, (real data(3 channels) + generated data(3 channels))]
    in_out_shape = (cfg.image_height, cfg.image_width, 3 * 2)
    all_input = Input(in_out_shape)

    # Split Input Data
    data_input = Lambda(lambda x: x[:, :, :, 0:3], output_shape=(cfg.image_height, cfg.image_width, 3))(all_input)
    generator_input = Lambda(lambda x: x[:, :, :, 3:6], output_shape=(cfg.image_height, cfg.image_width, 3))(all_input)

    # use same autoencoder(weights are shared)
    data_output = autoencoder(data_input)  # (bs, row, col, ch)
    generator_output = autoencoder(generator_input)

    # concatenate output to be same shape of input
    all_output = Concatenate(axis=-1)([data_output, generator_output])

    discriminator = DiscriminatorModel(all_input, all_output, name="discriminator")
    return discriminator

def conv_enc(x, filters, strides=(1, 1), name=None, n_layer=2):
    for i in range(1, n_layer):
        x = Convolution2D(filters, (3, 3), activation="elu", padding="same", name="%s/Conv%d" % (name, i))(x)
    # down sampling using striding
    x = Convolution2D(filters, (3, 3), activation="elu", padding="same", strides=strides,
                      name="%s/Conv%d" % (name, n_layer))(x)
    return x

def conv_dec(x, filters, upsample=None, name=None, n_layer=2):
    for i in range(1, n_layer+1):
        x = Convolution2D(filters, (3, 3), activation="elu", padding="same", name="%s/Conv%d" % (name, i))(x)
    if upsample:
        x = UpSampling2D()(x)
    return x

class DiscriminatorModel(Model):
    """Model which collects updates from loss_func.updates"""

    @property
    def updates(self):
        updates = super(DiscriminatorModel, self).updates
        if hasattr(self, 'loss_functions'):
            for loss_func in self.loss_functions:
                if hasattr(loss_func, 'updates'):
                    updates += loss_func.updates
        print('>>> updates!')
        return updates
```

```python
# generate some images with current generator's params
def generate(cfg, generator=None, num_image = 8):
    if generator is None:
        generator = build_generator(cfg)

    np.random.seed(999)
    z = np.random.uniform(-1, 1, cfg.emb_size * num_image)
    z = z.reshape((num_image, cfg.emb_size))
    x = generator.predict(z)
    images = np.clip(x * 255, 0, 255).astype("uint8")

    f, axes = plt.subplots(1, num_image)
    for i, a in enumerate(axes):
        image_x = images[i]
        img = Image.fromarray(image_x)
        a.imshow(img) #, interpolation='nearest')
        a.axis('off')
    plt.show()
```

```python
def create_generator_loss(autoencoder):
    def generator_loss(y_true, y_pred):
        y_pred_dash = autoencoder(y_pred)
        return K.mean(K.abs(y_pred - y_pred_dash), axis=[1, 2, 3])
    return generator_loss

class DiscriminatorLoss:
    __name__ = 'discriminator_loss'

    def __init__(self, k_0=0, lambda_k=0.001, gamma=0.5):
        self.lambda_k = lambda_k
        self.gamma = gamma
        self.k_var = K.variable(k_0, dtype=K.floatx(), name="discriminator_k")
        self.m_global_var = K.variable(0, dtype=K.floatx(), name="m_global")
        self.loss_real_x_var = K.variable(0, name="loss_real_x")  # for observation
        self.loss_gen_x_var = K.variable(0, name="loss_gen_x")    # for observation
        self.updates = []

    def __call__(self, y_true, y_pred):  # y_true, y_pred shape: (BS, row, col, ch * 2)
        data_true, generator_true = y_true[:, :, :, 0:3], y_true[:, :, :, 3:6]
        data_pred, generator_pred = y_pred[:, :, :, 0:3], y_pred[:, :, :, 3:6]
        loss_data = K.mean(K.abs(data_true - data_pred), axis=[1, 2, 3])
        loss_generator = K.mean(K.abs(generator_true - generator_pred), axis=[1, 2, 3])
        ret = loss_data - self.k_var * loss_generator

        # for updating values in each epoch, use `updates` mechanism
        # DiscriminatorModel collects Loss Function's updates attributes
        mean_loss_data = K.mean(loss_data)
        mean_loss_gen = K.mean(loss_generator)

        # update K
        new_k = self.k_var + self.lambda_k * (self.gamma * mean_loss_data - mean_loss_gen)
        new_k = K.clip(new_k, 0, 1)
        self.updates.append(K.update(self.k_var, new_k))

        # calculate M-Global
        m_global = mean_loss_data + K.abs(self.gamma * mean_loss_data - mean_loss_gen)
        self.updates.append(K.update(self.m_global_var, m_global))

        # let loss_real_x mean_loss_data
        self.updates.append(K.update(self.loss_real_x_var, mean_loss_data))

        # let loss_gen_x mean_loss_gen
        self.updates.append(K.update(self.loss_gen_x_var, mean_loss_gen))

        return ret

    @property
    def k(self):
        return K.get_value(self.k_var)

    @property
    def m_global(self):
        return K.get_value(self.m_global_var)

    @property
    def loss_real_x(self):
        return K.get_value(self.loss_real_x_var)

    @property
    def loss_gen_x(self):
        return K.get_value(self.loss_gen_x_var)
```

```python
def info(msg):
    now = datetime.now()
    print("%s: %s" % (now, msg))

def print_model_updates(model):
    training_updates = model.optimizer.get_updates(
        model._collected_trainable_weights,
        model.constraints,
        model.total_loss)
    updates = model.updates + training_updates
    print("\n".join(sorted([str(x[0]) for x in updates])))
```

```python
def training(cfg, epochs=3):
    # ensure dataset has only values in 0..1
    dataset = seg_dataset / 255.0

    batch_size = cfg.batch_size

    # building model and loading weight(if exists)
    autoencoder, generator, discriminator = build_model(cfg)

    loss_d = DiscriminatorLoss(cfg.k_0)  
    discriminator.compile(optimizer=Adam(), loss=loss_d)
    generator.compile(optimizer=Adam(), 
                      loss=create_generator_loss(autoencoder))
    lr_decay_step = 0
    last_m_global = np.Inf

    # print("Generator Update Variables")
    # print_model_updates(generator)
    #
    # print("Discriminator Update Variables")
    # print_model_updates(discriminator)

    for ep in range(1, epochs+1):
        np.random.seed(ep * 100)
        # generate Z layer values for discriminator and generator
        zd = np.random.uniform(-1, 1, (len(dataset), cfg.emb_size))
        zg = np.random.uniform(-1, 1, (len(dataset), cfg.emb_size))

        # shuffle dataset index
        index_order = np.arange(len(dataset))
        np.random.shuffle(index_order)

        # set Learning Rate
        lr = max(cfg.lr_0 * (cfg.lr_decay_rate ** lr_decay_step), 
                 cfg.min_lr)
        K.set_value(generator.optimizer.lr, lr)
        K.set_value(discriminator.optimizer.lr, lr)
        m_global_history = []
        info("LearningRate=%.7f" % lr)
        batch_len = len(dataset)//batch_size
        
        # Generate Image in each epoch for fun
        generate(cfg, generator)

        for b in range(batch_len):
            idxs = index_order[b*batch_size:(b+1)*batch_size]

            # training discriminator
            in_x1 = dataset[idxs]  # (bs, row, col, ch)
            in_x2 = generator.predict_on_batch(zd[idxs])
            in_x = np.concatenate([in_x1, in_x2], axis=-1)  # (bs, row, col, ch*2)
            loss_discriminator = discriminator.train_on_batch(in_x, in_x)

            # training generator
            in_x1 = zg[idxs]
            loss_generator = generator.train_on_batch(in_x1, np.zeros_like(in_x2))  # y_true is meaningless

            # record M-Global
            m_global_history.append(loss_d.m_global)
            print("\r{}%".format(100 * b // batch_len), end="") 
            sys.stdout.flush()
            
        log_info = dict(
            epoch=ep,
            m_global=loss_d.m_global,
            loss_discriminator=loss_discriminator,
            loss_generator=loss_generator,
            loss_real_x=loss_d.loss_real_x,
            loss_gen_x=loss_d.loss_gen_x,
            k=loss_d.k,
            lr=lr,
        )
        info("ep=%(epoch)s, MGlobal=%(m_global).5f, "
             "Loss(D)=%(loss_discriminator).5f, Loss(G)=%(loss_generator).5f, Loss(X)=%(loss_real_x).5f, "
             "Loss(G(Zd))=%(loss_gen_x).5f, K=%(k).6f" % log_info)

        m_global = np.average(m_global_history)
        if last_m_global <= m_global:  # decay LearningRate
            lr_decay_step += 1
        last_m_global = m_global
```

```python
cfg = BEGANConfig()
try:
    training(cfg, epochs=500)
    print('Done.')
except KeyboardInterrupt:
    print('Interrupted, as you wish...')
```

## Mais desafios para GANs

<img src="images/gan_challenges.png" alt="GAN challenges" style="width: 400px;"/>

* _Contagem_: em imagens, é comum que a GAN gere múltiplas ocorrências de atributos de forma a não ser natural. Por exemplo, gerar um cão com mais de dois olhos, múltiplas pernas ou cabeças. Uma forma de eliminar este problema é dar imagens homogêneas no treino. Mas há bastante pesquisa em compreensão do espaço de representação;
* _Perspectiva_: a GAN tem dificuldade em compreender a natureza 3-D do mundo em imagens 2-D, de forma a aplicar o que aprendeu com sucesso em novas imagens. Relacionada a esta linha, há muitos trabalhos sobre GANs que operam com dados tridimensionais, mas ainda com resultados pouco encorajadores;
* _Estrutura Global_: a GAN não compreende a imagem de um objeto complexo como uma estrutura global complexa que precisa ser preservada. Como o anterior, isto motiva mais pesquisa em representação 3D com GANs. 

    Outra linha de pesquisa consiste em aprender representações abstratas de conceitos e suas estruturas (ver https://research.googleblog.com/2017/04/teaching-machines-to-draw.html). Neste estudo que não é baseado em uma GAN, uma seq2seq com inferência variacional é usada para aprender a desenhar sketches. No exemplo abaixo, a rede traduz entradas humanas para sketches de porquinhos. Note como a rede entende a estrutura geral de um porco:
